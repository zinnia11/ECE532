1) A binary classifier decides whether data is from one of two classes, labeled 1 and -1.In this problem the data is described by two features $x_1$ and $x_2$ and the classificationdecision is made as follows. The class with label 1 is decided if $x_1a_1 + x_2a_2 > b$ whilethe class with label -1 is decided when $x_1a_1 + x_2a_2 < b$. Here $a_1$, $a_2$, and $b$ are givenreal numbers.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# number of features
p = 2

# number of examples
n = 5000

# generate matrix of n (random) examples of p features with a column of all ones
X0 = np.random.rand(n,p)
onevec = np.ones(shape = (n,1))
X = np.concatenate((X0,onevec),axis=1)

# Classifier weights
w = [[-1], [2], [-0.4]]

# Multiply feature matrix with weights  yhat = X*w
yhat = X@w

# Decide which class based on whether yhat is >< 0
# sign function returns +1 when yhat(i)>0 and -1 when yhat(i)<0
pred_label = np.sign(yhat);

plt.scatter(X[:,[0]],X[:,[1]], color = ['r' if i==-1 else 'b' for i in pred_label])

plt.xlabel("x_1")
plt.ylabel("x_2")
plt.title("Case 2")
plt.axis('square')
plt.show()
```
